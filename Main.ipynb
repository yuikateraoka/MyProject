```python
import sys
import MeCab

import urllib

from collections import defaultdict
from gensim import corpora
from gensim import models

import pandas as pd
```

```python

#!/usr/bin/env python
"""
Test of MeCab library
"""

# MeCab による単語への分割関数 (名詞のみ残す)


def split_text_only_noun(text):
    tagger = MeCab.Tagger()

    words = []
    for c in tagger.parse(text).splitlines()[:-1]:
        surface, feature = c.split('\t')
#         chunks = line.split('\t')
        pos = feature.split(',')[0]
        if pos == '名詞':
            words.append(surface)
    return ' '.join(words)
```

```python
book = open(R"DATA.txt", "rt", encoding="utf-8")
text = book.read()
text_corpus = text.split()
# print(text)
```

```python
split_text = []
split_text = split_text_only_noun(text)
split_text = split_text.split()
```

```python

 req = urllib.request.Request('http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt')
 
with urllib.request.urlopen(req) as res:
    stopwords = res.read().decode('utf-8').split('\r\n')
 
while '' in stopwords:
    stopwords.remove('')
    t=[]
    t = ['→', '氏','型','EOS','空き家','事例','空き','場所','活用','街','利用','の','ん','階','■','1','2','【','】','☟','☟','●','▼・ᴥ・▼','⇒','‍♂','Ф','°','①,','お気に入り','まとめ',r'[!-~]',r'[︰-＠]','\n','②','④','♪']
    
    
    stopword = stopwords+t
```

```python


m = MeCab.Tagger("-Ochasen")

stoplist = stopword

a = [[i.split()[0] for i in m.parse(document.lower()).splitlines()] for document in split_text] 
texts = [[word for word in document if word not in stoplist] for document in a]  
```

```python

frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1
        
processed_corpus = [[token for token in text if frequency[token] > 1] for text in texts]
dictionary = corpora.Dictionary(processed_corpus)
```

```python
bow_corpus = [dictionary.doc2bow(text) for text in processed_corpus]

tfidf = models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]
for doc in corpus_tfidf:
    print(doc)
```

```python
N = 4
lda_model = models.LdaModel(bow_corpus, id2word=dictionary, num_topics= N)
corpus_lda = lda_model[bow_corpus] 
for doc, as_text in zip(corpus_lda, text_corpus):
     print(doc, as_text)
```
